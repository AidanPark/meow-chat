```python
def preprocess(
    img_bytes: bytes,    
    debug: bool = True,
) -> bytes:
    try:
        if debug:
            print("ğŸ¥ ì˜ë£Œ ë¬¸ì„œìš© ì „ì²˜ë¦¬ ì‹œì‘")
        
        img = open_with_exif(img_bytes) # EXIF íšŒì „ êµì •
        original_size = img.size
        
        if debug:
            print(f"ğŸ“ ì›ë³¸ ì´ë¯¸ì§€ í¬ê¸°: {original_size}")
        
        steps = [
            # ê¸°ë³¸ ì •ë¦¬ (í•„ìˆ˜ ì„ í–‰ ë‹¨ê³„)
            (flatten_transparency, {}),              # íˆ¬ëª…ë„ ì œê±°
            (normalize_mode, {}),                    # ëª¨ë“œ í‘œì¤€í™” (L/RGB)
        
            # í•´ìƒë„ í‘œì¤€í™” (ìµœì†Œ í’ˆì§ˆ í™•ë³´)
            (upscale_min_resolution, {}),            
            
            # ë¹›ë°˜ì‚¬ ì œê±°
            (suppress_glare, {}),                   
            
            # ëŒ€ë¹„ í–¥ìƒ
            (weak_autocontrast, {}),               
            
            # ìƒ‰ í…ìŠ¤íŠ¸ ë³´ì •(ìˆìœ¼ë©´)
            (blacken_reddish_text, {}),  
            (blacken_bluish_text, {}), 

            # íšŒìƒ‰í™”
            (to_grayscale, {}),    

            # ê¸°í•˜í•™ì  ë³´ì • (í”½ì…€ ìˆ˜ì¤€ ì‘ì—…)
            # (auto_crop_with_margin, {}),           # ì—¬ë°± ì œê±°
            (conditional_dewarp, {}),                # í˜ì´ì§€ íœ˜ì–´ì§ ë³´ì • 
            (deskew_textlines, {}),                  # ë¯¸ì„¸ ê¸°ìš¸ê¸° ë³´ì •               

            # í‘œ ë¼ì¸ ê°•í™”
            (enhance_table_lines, {}), 
            
            # ìµœì¢… í¬ê¸° ì¡°ì • (í† í° ì ˆì•½)
            (downscale_target_long_edge, {}),      

            # ì„ ëª…ë„ í–¥ìƒ
            (conservative_sharpen, {}),           
        ]
    
        # íŒŒì´í”„ë¼ì¸ ì‹¤í–‰
        img = apply_pipeline(img, steps)
        
        # PNGë¡œ ì €ì¥
        buf = io.BytesIO()
        if img.mode == 'L':  # ê·¸ë ˆì´ìŠ¤ì¼€ì¼
            img.save(buf, format="JPEG", quality=85, optimize=True)
        else:  # RGB
            img.save(buf, format="JPEG", quality=85, optimize=True, subsampling=2)
        result_bytes = buf.getvalue()
        
        if debug:
            final_size = img.size
            print(f"ğŸ“ ìµœì¢… ì´ë¯¸ì§€ í¬ê¸°: {final_size}")
            print(f"ğŸ”„ í¬ê¸° ë³€í™”: {original_size} â†’ {final_size}")
            # ìš©ëŸ‰ ë³€í™” ë° í¼ì„¼íŠ¸ ê³„ì‚°
            original_bytes = len(img_bytes)
            result_bytes_len = len(result_bytes)
            percentage = (result_bytes_len / original_bytes) * 100
            
            print(f"ğŸ’¾ ìš©ëŸ‰ ë³€í™”: {original_bytes:,} bytes â†’ {result_bytes_len:,} bytes ({percentage:.1f}%)")
            print("âœ… ì „ì²˜ë¦¬ ì™„ë£Œ")
        
        return result_bytes
        
    except Exception as e:
        if debug:
            print(f"âŒ ì „ì²˜ë¦¬ ì‹¤íŒ¨: {e}")
            import traceback
            traceback.print_exc()
        
        print(f"âš ï¸ ì „ì²˜ë¦¬ ì‹¤íŒ¨, ì›ë³¸ ë°˜í™˜: {e}")
        return img_bytes 
```

```python
import re
import json
from typing import List, Dict, Any, Optional, Tuple

# LLM ì „ìš© íŒŒì„œ: LLMì´ í† í° ì¸ë±ìŠ¤ë§Œ ë°˜í™˜í•˜ê³ , ì‹¤ì œ ê°’ì€ ë°˜ë“œì‹œ ocr_dataì˜ í…ìŠ¤íŠ¸/ì»¨í”¼ë˜ìŠ¤ë¡œë¶€í„° ì¬êµ¬ì„±í•©ë‹ˆë‹¤.
# - valueëŠ” í•­ìƒ ìˆ˜ì¹˜í˜•ì´ì–´ì•¼ í•˜ë©°, ê°’ ëì˜ H/L/N/High/Low/Normal ì ‘ë¯¸ëŠ” ì œê±° í›„ ìˆ«ìë§Œ ì¶”ì¶œí•©ë‹ˆë‹¤.
# - code_confidenceëŠ” ocr_dataì˜ í•´ë‹¹ code í† í°ì˜ confidenceë¥¼ ë¬¸ìì—´ë¡œ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤(ë°˜ì˜¬ë¦¼/í¬ë§· ë³€ê²½ ì—†ìŒ).
# - reference ì—­ì‹œ ocr_dataì˜ í…ìŠ¤íŠ¸ë¡œë¶€í„°ë§Œ ì¶”ì¶œ/ê°€ê³µí•©ë‹ˆë‹¤(ì„ì˜ ìƒì„± ê¸ˆì§€).

_num_re = re.compile(r"[-+]?\d+(?:\.\d+)?")
_range_re = re.compile(r"([-+]?\d+(?:\.\d+)?)\s*[-â€“~]\s*([-+]?\d+(?:\.\d+)?)")
_suffix_re = re.compile(r"\s*(?:\(|\[)?\s*(?i:(?:high|low|normal)|h|l|n)\s*(?:\)|\])?\s*$")

# ê°„ë‹¨ ë‹¨ìœ„ ì‚¬ì „(ì´ˆê¸° 5ê°œ, ì´í›„ í™•ì¥ ì˜ˆì •)
_CANON_UNITS = {
    "%",
    "g/dl",
    "mg/dl",
    "u/l",
    "mmol/l",
}


def _normalize_str(s: str) -> str:
    if not s:
        return ""
    t = s.strip().lower()
    # ìœ ë‹ˆì½”ë“œ ë³€í˜• ìµœì†Œ í†µì¼
    t = t.replace("Î¼", "u").replace("Âµ", "u")
    t = re.sub(r"\s+", "", t)
    return t


def _looks_numeric(text: str) -> bool:
    return bool(_num_re.search(text or ""))


def _looks_unit_like(text: str) -> bool:
    if not text:
        return False
    t = _normalize_str(text)
    if t in _CANON_UNITS:
        return True
    # ê°„ë‹¨ íœ´ë¦¬ìŠ¤í‹±: % í¬í•¨ ë˜ëŠ” '/' í¬í•¨(ë‹¨, ìˆœìˆ˜ ìˆ«ì/ë‚ ì§œ íŒ¨í„´ ì œì™¸)
    if "%" in t:
        return True
    if "/" in t and not re.fullmatch(r"\d{1,4}/\d{1,2}/\d{1,2}", t):
        return True
    # ì§€ìˆ˜/ë‹¨ìœ„ í”ì 
    if any(ch in t for ch in ["^", "Ã—", "x10"]):
        return True
    return False


def _extract_numeric(text: str) -> Optional[str]:
    if not text:
        return None
    t = text.strip()
    # ì ‘ë¯¸(H/L/N/High/Low/Normal) ì œê±°
    t = _suffix_re.sub("", t)
    m = _num_re.search(t)
    return m.group(0) if m else None


def _extract_range(text: str) -> Tuple[Optional[str], Optional[str]]:
    if not text:
        return None, None
    m = _range_re.search(text)
    if not m:
        return None, None
    return m.group(1), m.group(2)


def _safe_get_token(tokens: List[Dict[str, Any]], idx: Optional[int]) -> Optional[Dict[str, Any]]:
    if idx is None:
        return None
    if not isinstance(idx, int):
        return None
    if idx < 0 or idx >= len(tokens):
        return None
    return tokens[idx]


def _compute_doc_y_bounds(polys: List[List[List[float]]]) -> Tuple[float, float]:
    ys = []
    for poly in polys or []:
        for x, y in poly:
            ys.append(float(y))
    if not ys:
        return 0.0, 1.0
    return min(ys), max(ys) if max(ys) > min(ys) else (min(ys), min(ys) + 1.0)


def _y_box_from_poly(poly: List[List[float]]) -> Tuple[float, float]:
    y_vals = [float(p[1]) for p in poly]
    return min(y_vals), max(y_vals)


def _attach_y_features(
    ocr_items: List[Dict[str, Any]],
    layout_page: Optional[Dict[str, Any]]
) -> List[Dict[str, Any]]:
    """
    ocr_items: [{text, confidence}, ...] (post-processed tokens)
    layout_page: result[0] (dict with rec_texts, rec_scores, rec_polys)
    Returns: list aligned to filtered ocr_items order with y_top_norm, y_bottom_norm, h_norm
    Mapping: greedy by substring equivalence on normalized texts; fallback to nearest index.
    """
    n = len(ocr_items)
    feats = [{"y_top_norm": None, "y_bottom_norm": None, "h_norm": None} for _ in range(n)]

    if not layout_page or not isinstance(layout_page, dict):
        return feats

    rec_texts = layout_page.get("rec_texts", []) or []
    rec_polys = layout_page.get("rec_polys", layout_page.get("dt_polys", [])) or []
    if not rec_texts or not rec_polys or len(rec_texts) != len(rec_polys):
        return feats

    # ë¬¸ì„œ ë†’ì´ êµ¬í•´ ì •ê·œí™” ì¤€ë¹„
    doc_y_min, doc_y_max = _compute_doc_y_bounds(rec_polys)
    doc_h = max(doc_y_max - doc_y_min, 1e-6)

    # ì›ì‹œ ë¦¬ìŠ¤íŠ¸ ì „ì²˜ë¦¬
    raw_norm_texts = [_normalize_str(t) for t in rec_texts]
    raw_y_boxes = []  # (y_top, y_bottom)
    for poly in rec_polys:
        y0, y1 = _y_box_from_poly(poly)
        raw_y_boxes.append((y0, y1))

    j = 0  # greedy pointer over raw tokens
    for i, item in enumerate(ocr_items):
        t = _normalize_str(str(item.get("text", "")))
        matched_k = None
        for k in range(j, len(raw_norm_texts)):
            r = raw_norm_texts[k]
            if not r:
                continue
            if t == r or (t and t in r) or (r and r in t):
                matched_k = k
                break
        if matched_k is None:
            # fallback: stick to pointer if valid, else last index
            matched_k = min(j, len(raw_norm_texts) - 1)
        # update pointer to avoid reusing too far back
        j = max(j, matched_k)

        y0, y1 = raw_y_boxes[matched_k]
        y_top_norm = (y0 - doc_y_min) / doc_h
        y_bottom_norm = (y1 - doc_y_min) / doc_h
        h_norm = max(y_bottom_norm - y_top_norm, 0.0)
        feats[i] = {
            "y_top_norm": y_top_norm,
            "y_bottom_norm": y_bottom_norm,
            "h_norm": h_norm,
        }

    return feats


def _assign_line_ids(y_feats: List[Dict[str, Optional[float]]], alpha: float = 0.7) -> Tuple[List[int], List[int]]:
    """
    ì„ê³„ ê¸°ë°˜ 1D í´ëŸ¬ìŠ¤í„°ë§: center_y ê°„ê²©ì´ tau ì´í•˜ë©´ ê°™ì€ í–‰.
    tau = alpha * median(h_norm). h_normì´ ë¶€ì¬ ì‹œ 0.02 ê¸°ë³¸.
    Returns: (line_id_list, line_rank_list)
    """
    n = len(y_feats)
    if n == 0:
        return [], []

    centers = []
    h_list = []
    for i, f in enumerate(y_feats):
        y_top = f.get("y_top_norm")
        y_bottom = f.get("y_bottom_norm")
        h = f.get("h_norm")
        if y_top is None or y_bottom is None:
            centers.append((i, None))
        else:
            centers.append((i, (y_top + y_bottom) / 2.0))
        if h is not None and h > 0:
            h_list.append(h)

    if not h_list:
        median_h = 0.02
    else:
        h_sorted = sorted(h_list)
        mid = len(h_sorted) // 2
        median_h = (h_sorted[mid] if len(h_sorted) % 2 == 1 else (h_sorted[mid - 1] + h_sorted[mid]) / 2.0)
    tau = max(alpha * median_h, 1e-3)

    # ì •ë ¬ëœ ì¸ë±ìŠ¤ì— ëŒ€í•´ ë¼ì¸ ë¶€ì—¬
    sortable = [(i, c) for (i, c) in centers if c is not None]
    missing = [i for (i, c) in centers if c is None]
    sortable.sort(key=lambda x: x[1])

    line_id_by_idx = [-1] * n
    current_line = -1
    prev_c = None
    for idx, c in sortable:
        if prev_c is None or (c - prev_c) > tau:
            current_line += 1
        line_id_by_idx[idx] = current_line
        prev_c = c

    # ëˆ„ë½(center ì—†ìŒ)ì€ ê°€ì¥ ê°€ê¹Œìš´ ì´ì›ƒ ë¼ì¸ìœ¼ë¡œ ë¶™ì´ê±°ë‚˜ 0ìœ¼ë¡œ
    for idx in missing:
        # ê°„ë‹¨íˆ 0ìœ¼ë¡œ
        line_id_by_idx[idx] = max(current_line, 0)

    # line_rankëŠ” line_idì™€ ë™ì¼(0..L-1)
    return line_id_by_idx, line_id_by_idx


def _build_tests_from_indices(tokens: List[Dict[str, Any]], plan: Dict[str, Any]) -> Dict[str, Any]:
    out = []
    tests = plan.get("tests", []) if isinstance(plan, dict) else []
    for t in tests:
        code_idx = t.get("code_idx")
        value_idx = t.get("value_idx")
        unit_idx = t.get("unit_idx")
        ref_min_idx = t.get("ref_min_idx")
        ref_max_idx = t.get("ref_max_idx")
        range_idx = t.get("range_idx")

        tok_code = _safe_get_token(tokens, code_idx)
        tok_value = _safe_get_token(tokens, value_idx)
        tok_unit = _safe_get_token(tokens, unit_idx)
        tok_ref_min = _safe_get_token(tokens, ref_min_idx)
        tok_ref_max = _safe_get_token(tokens, ref_max_idx)
        tok_range = _safe_get_token(tokens, range_idx)

        code_text = (tok_code or {}).get("text") or "UNKNOWN"
        code_confidence = str((tok_code or {}).get("confidence") if tok_code and "confidence" in tok_code else "")

        # value: ë°˜ë“œì‹œ ìˆ˜ì¹˜í˜•, ì ‘ë¯¸(H/L/N/High/Low/Normal) ì œê±° í›„ ìˆ«ìë§Œ
        raw_value_text = (tok_value or {}).get("text") if tok_value else None
        value_num = _extract_numeric(raw_value_text or "")
        if value_num is None:
            value_num = "UNKNOWN"
        # value_confidence: value í† í°ì˜ confidence ê·¸ëŒ€ë¡œ ë¬¸ìì—´í™”
        value_confidence = str((tok_value or {}).get("confidence") if tok_value and "confidence" in tok_value else "")

        # unit: ê³µë°±ìœ¼ë¡œ ë‚˜ë‰˜ëŠ” ê²½ìš° ë§ˆì§€ë§‰ í† í°ì„ ë‹¨ìœ„ë¡œ í•´ì„ (unit_idxì—ì„œë§Œ ì¶”ì¶œ)
        unit_text = "UNKNOWN"
        unit_confidence = "UNKNOWN"
        if tok_unit:
            ut = (tok_unit or {}).get("text") or ""
            ut = ut.strip()
            if ut:
                unit_text = ut.split()[-1]
            unit_confidence = str((tok_unit or {}).get("confidence") if "confidence" in tok_unit else "") or "UNKNOWN"

        # reference: ref_min/ref_max ì¸ë±ìŠ¤ê°€ ìš°ì„ , ì—†ìœ¼ë©´ range_idxì—ì„œ íŒŒì‹±
        ref_min = _extract_numeric((tok_ref_min or {}).get("text") or "") if tok_ref_min else None
        ref_max = _extract_numeric((tok_ref_max or {}).get("text") or "") if tok_ref_max else None
        ref_min_source = "min" if (tok_ref_min and ref_min is not None) else None
        ref_max_source = "max" if (tok_ref_max and ref_max is not None) else None
        if (ref_min is None or ref_max is None) and tok_range:
            rmin, rmax = _extract_range((tok_range or {}).get("text") or "")
            if ref_min is None and rmin is not None:
                ref_min = rmin
                ref_min_source = "range"
            if ref_max is None and rmax is not None:
                ref_max = rmax
                ref_max_source = "range"
        # ìµœì¢… ê°’ ë³´ì • ë° confidence ë§¤í•‘
        if ref_min is None:
            ref_min = "UNKNOWN"
        if ref_max is None:
            ref_max = "UNKNOWN"
        if ref_min_source == "min" and tok_ref_min:
            ref_min_confidence = str((tok_ref_min or {}).get("confidence"))
        elif ref_min_source == "range" and tok_range:
            ref_min_confidence = str((tok_range or {}).get("confidence"))
        else:
            ref_min_confidence = "UNKNOWN"
        if ref_max_source == "max" and tok_ref_max:
            ref_max_confidence = str((tok_ref_max or {}).get("confidence"))
        elif ref_max_source == "range" and tok_range:
            ref_max_confidence = str((tok_range or {}).get("confidence"))
        else:
            ref_max_confidence = "UNKNOWN"

        out.append({
            "code": str(code_text),
            "code_confidence": code_confidence if code_confidence != "" else "UNKNOWN",
            "value": str(value_num),
            "value_confidence": value_confidence if value_confidence != "" else "UNKNOWN",
            "reference_min": str(ref_min),
            "reference_min_confidence": ref_min_confidence,
            "reference_max": str(ref_max),
            "reference_max_confidence": ref_max_confidence,
            "unit": str(unit_text),
            "unit_confidence": unit_confidence if unit_confidence != "" else "UNKNOWN",
        })
    return {"tests": out}


def analyze_medical_ocr_data_with_llm(
    ocr_data: List[Dict[str, Any]],
    model: str = "gpt-4.1-mini",
    seed: Optional[int] = 42,
    layout_page: Optional[Dict[str, Any]] = None,
    verbose: bool = False,
    return_plan: bool = False,
) -> Any:
    """
    LLMë§Œì„ ì‚¬ìš©í•´ ë‹¤ì–‘í•œ í˜•íƒœì˜ OCR í† í°ìœ¼ë¡œë¶€í„° í…Œì´ë¸”(ê²€ì‚¬ì½”ë“œ/ê°’/ì°¸ê³ ë²”ìœ„/ë‹¨ìœ„)ì„ ì¶”ì¶œí•©ë‹ˆë‹¤.
    - LLMì€ ë°˜ë“œì‹œ tokens ë°°ì—´ì˜ ì¸ë±ìŠ¤ë§Œ ë°˜í™˜í•©ë‹ˆë‹¤(ë¬¸ìì—´ ê°’ì„ ì§ì ‘ ìƒì„± ê¸ˆì§€).
    - valueëŠ” ìˆ˜ì¹˜í˜•ë§Œ í—ˆìš©í•˜ë©°, ì ‘ë¯¸ H/L/N/High/Low/Normalì€ ë¬´ì‹œ í›„ ìˆ«ìë§Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
    - code_confidenceëŠ” ocr_dataì˜ í•´ë‹¹ code í† í° confidenceë¥¼ ë¬¸ìì—´ë¡œ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.
    - y ì „ìš© ì‹ í˜¸: y_top_norm, y_bottom_norm, h_norm, line_id, line_rankë¥¼ ì œê³µí•˜ê³  x/wëŠ” ì œê³µí•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.
    - model: ì‚¬ìš©í•  OpenAI ëª¨ë¸ ì´ë¦„(ê¸°ë³¸ê°’ "gpt-4.1-mini").
    - seed: ê°€ëŠ¥í•˜ë©´ ë™ì¼ ê²°ê³¼ ì¬í˜„ì„ ìœ„í•œ ì‹œë“œ(ëª¨ë¸/SDKê°€ ì§€ì›í•˜ëŠ” ê²½ìš°ì—ë§Œ ì ìš©).
    - verbose: ë‚´ë¶€ ì§„ë‹¨ ì¶œë ¥(tokens_for_llm/plan) í‘œì‹œ ì—¬ë¶€.
    - return_plan: Trueë©´ (plan, ê²°ê³¼êµ¬ì¡°) íŠœí”Œì„ ë°˜í™˜í•©ë‹ˆë‹¤.
    ë°˜í™˜: {"tests":[{...}]} ë˜ëŠ” (plan, {"tests": [...]})
    """
    # 1) í…ìŠ¤íŠ¸ê°€ ìˆëŠ” í† í°ë§Œ í•„í„°ë§ (ê¸°ì¡´ ë™ì‘ ìœ ì§€)
    filtered = [x for x in ocr_data if str(x.get("text", "")).strip()]

    # 2) y íŠ¹ì„± ë¶€ì°© (layout_page ê¸°ë°˜ ë§¤í•‘)
    y_feats = _attach_y_features(filtered, layout_page)

    # 3) line_id/line_rank ê³„ì‚° (y-only)
    line_ids, line_ranks = _assign_line_ids(y_feats, alpha=0.7)

    # 4) tokens_for_llm êµ¬ì„± (x/w ì œì™¸, ì„ íƒ íŒíŠ¸ í¬í•¨)
    tokens_for_llm = []
    for i, x in enumerate(filtered):
        y_top_norm = y_feats[i].get("y_top_norm")
        y_bottom_norm = y_feats[i].get("y_bottom_norm")
        h_norm = y_feats[i].get("h_norm")
        txt = str(x.get("text", ""))
        cf = float(x.get("confidence", 0.0))
        tokens_for_llm.append({
            "idx": i,
            "text": txt,
            "confidence": cf,
            "y_top_norm": y_top_norm,
            "y_bottom_norm": y_bottom_norm,
            "h_norm": h_norm,
            "line_id": int(line_ids[i]) if i < len(line_ids) and line_ids[i] is not None else 0,
            "line_rank": int(line_ranks[i]) if i < len(line_ranks) and line_ranks[i] is not None else 0,
            # ê°€ë²¼ìš´ íŒíŠ¸(ì„ íƒ)
            "value_like": _looks_numeric(txt),
            "unit_like": _looks_unit_like(txt),
        })

    if verbose:
        print(f"tokens_for_llm: {len(tokens_for_llm)} items (y-only features attached)")

    system_prompt = """
ë‹¹ì‹ ì€ ë°˜ë ¤ë™ë¬¼ ê²€ì‚¬ê²°ê³¼ì§€ OCR í† í°ë“¤ë¡œë¶€í„° í…Œì´ë¸” ë°ì´í„°ë¥¼ êµ¬ì¡°í™”í•˜ëŠ” ì „ë¬¸ê°€ì…ë‹ˆë‹¤.
ë‹¤ìŒ ì§€ì¹¨ì„ ë°˜ë“œì‹œ ì§€í‚¤ì„¸ìš”:
1) í–‰ ê·¸ë£¹í™”: ê°™ì€ line_idëŠ” ê°™ì€ í–‰ìœ¼ë¡œ ê°„ì£¼í•©ë‹ˆë‹¤. x/w ì •ë³´ëŠ” ì œê³µí•˜ì§€ ì•Šìœ¼ë¯€ë¡œ ì‚¬ìš©í•˜ì§€ ë§ˆì„¸ìš”.
2) ê°’/ë‹¨ìœ„/ì°¸ê³ ë²”ìœ„ ë§¤ì¹­ì€ ë°˜ë“œì‹œ ê°™ì€ line_id ë‚´ì—ì„œë§Œ ì°¾ìŠµë‹ˆë‹¤. ë‹¤ë¥¸ í–‰(line_idê°€ ë‹¤ë¥¸ í† í°)ì€ ì ˆëŒ€ ì‚¬ìš©í•˜ì§€ ë§ˆì„¸ìš”.
3) ì°¸ê³ ë²”ìœ„ëŠ” 'min-max' í•˜ë‚˜ì˜ í† í°ì´ê±°ë‚˜, min/maxê°€ ë¶„ë¦¬ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ë“¤ë„ ê°™ì€ line_id ë‚´ì—ì„œë§Œ ì„ íƒí•˜ì„¸ìš”.
4) ë°˜í™˜ ì‹œ ì ˆëŒ€ ë¬¸ìì—´ ê°’(ì½”ë“œ/ê°’/ë‹¨ìœ„/ë²”ìœ„)ì„ ì§ì ‘ ìƒì„±í•˜ì§€ ë§ê³ , ì œê³µëœ tokens ë°°ì—´ì˜ ì¸ë±ìŠ¤ë§Œ ì‚¬ìš©í•˜ì„¸ìš”.
5) ê²°ê³¼ í¬ë§·: {"tests":[{"code_idx":int,"value_idx":int|null,"unit_idx":int|null,"ref_min_idx":int|null,"ref_max_idx":int|null,"range_idx":int|null}]}
6) ì¸ë±ìŠ¤ëŠ” tokens[idx]ë¥¼ ê°€ë¦¬ì¼œì•¼ í•˜ë©° ìœ íš¨í•˜ì§€ ì•Šì€ ì¸ë±ìŠ¤ë¥¼ ìƒì„±í•˜ì§€ ë§ˆì„¸ìš”.
"""

    user_prompt = """
ì•„ë˜ tokens(JSON ë°°ì—´, ê° ì›ì†ŒëŠ” {idx,text,confidence,y_top_norm,y_bottom_norm,h_norm,line_id,line_rank,value_like,unit_like})ë§Œì„ ì´ìš©í•´ ê²€ì‚¬ í•­ëª©ì„ ì¶”ì¶œí•˜ì„¸ìš”.
- ë¬¸ìì—´ ê°’ì„ ì§ì ‘ ë§Œë“¤ì§€ ë§ê³  ì¸ë±ìŠ¤ë§Œ ë°˜í™˜í•˜ì„¸ìš”.
- value/unit/referenceëŠ” ê°™ì€ line_id ë‚´ì—ì„œë§Œ ì„ íƒí•˜ì„¸ìš”. line_idê°€ ë‹¤ë¥´ë©´ ì„ íƒí•˜ì§€ ë§ê³  nullë¡œ ë‘ì„¸ìš”.
- unit_like=trueì¸ í† í°ì„ ìš°ì„  ë‹¨ìœ„ í›„ë³´ë¡œ ê³ ë ¤í•˜ì„¸ìš”. range íŒ¨í„´ì´ ìˆìœ¼ë©´ range_idxë¥¼ ì‚¬ìš©í•˜ì„¸ìš”(ë°˜ë“œì‹œ ê°™ì€ line_id).
"""

    # Pylance ì¹œí™”ì : contentë¥¼ ë‹¨ì¼ ë¬¸ìì—´ë¡œ í•©ì¹˜ê³ , seedëŠ” extra_bodyë¡œ ì „ë‹¬
    content_str = user_prompt + "\n" + json.dumps({"tokens": tokens_for_llm}, ensure_ascii=False)

    payload: Dict[str, Any] = {
        "model": model,
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": content_str},
        ],
        "temperature": 0,
        "response_format": {"type": "json_object"},
    }
    if seed is not None:
        payload["extra_body"] = {"seed": seed}

    response = client.chat.completions.create(**payload)

    plan = json.loads(response.choices[0].message.content)
    if verbose:
        print(f"plan {model} tests={len(plan.get('tests', []))}")

    result_struct = _build_tests_from_indices(tokens_for_llm, plan)
    return (plan, result_struct) if return_plan else result_struct

print("ğŸ§  LLM ì „ìš©(ì¸ë±ìŠ¤ ê¸°ë°˜) íŒŒì„œ: y-only line signals + strict same-line matching ì ìš©!")
```
