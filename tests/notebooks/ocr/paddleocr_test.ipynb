### Step 1. Îπà Í≤∞Í≥º Ï¥àÍ∏∞Ìôî

- ÏµúÏ¢Ö Í≤∞Í≥º Ïª®ÌÖåÏù¥ÎÑà `doc_result`Î•º ÏïÑÎûò ÌÇ§Î°ú Ï¥àÍ∏∞Ìôî
  - hospital_name, client_name, patient_name, inspection_date, tests([])

````python
from typing import Any, Dict, List
import json
import time

# Pipeline ÏãúÏûë ÏãúÍ∞Ñ Í∏∞Î°ù
pipeline_start_time = time.time()
print(f"‚è±Ô∏è Pipeline ÏãúÏûë ÏãúÍ∞Ñ: {time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(pipeline_start_time))}")

DocumentResult = Dict[str, Any]


def init_doc_result() -> DocumentResult:
    """ÌëúÏ§Ä Í≤∞Í≥º Ïª®ÌÖåÏù¥ÎÑà Ï¥àÍ∏∞Ìôî.
    Keys:
      - hospital_name: str | None
      - client_name: str | None
      - patient_name: str | None
      - inspection_date: str | None (YYYY-MM-DD)
      - tests: list[dict]
    """
    return {
        "hospital_name": None,
        "client_name": None,
        "patient_name": None,
        "inspection_date": None,
        "tests": [],
    }


def validate_doc_result(doc: DocumentResult) -> tuple[bool, str]:
    """ÏµúÏÜå Ïú†Ìö®ÏÑ± Í≤ÄÏ¶ù: tests Ï°¥Ïû¨ Ïó¨Î∂Ä.
    returns: (is_valid, message)
    """
    if not doc.get("tests"):
        return False, "tests ÎπÑÏñ¥ÏûàÏùå"
    return True, "ok"


# Ïã§Ìñâ ÏòàÏãú
_doc = init_doc_result()
print("‚úÖ Ï¥àÍ∏∞ÌôîÎêú Í≤∞Í≥º Ïª®ÌÖåÏù¥ÎÑà:")
print(json.dumps(_doc, ensure_ascii=False, indent=2))
print("Ïú†Ìö®ÏÑ±:", validate_doc_result(_doc))
````

````python
# ÎùºÏù∏Î≥Ñ ÌÖçÏä§Ìä∏Îßå ÌååÏù¥ÌîÑ(|) Íµ¨Î∂ÑÏúºÎ°ú Ï∂úÎ†•
def _tok_text(tok):
    if isinstance(tok, dict):
        t = tok.get("text")
        if isinstance(t, str):
            return t.strip()
    if isinstance(tok, str):
        return tok.strip()
    return str(tok).strip()

def _line_to_pipe(line):
    # ÏßÄÏõê ÌòïÌÉú:
    # - Î¶¨Ïä§Ìä∏/ÌäúÌîå: ÌÜ†ÌÅ∞ Î¶¨Ïä§Ìä∏ (Î¨∏ÏûêÏó¥/ÎîïÏÖîÎÑàÎ¶¨ ÌòºÌï© Í∞ÄÎä•)
    # - ÎîïÏÖîÎÑàÎ¶¨: 'tokens' | '_tokens' | 'line' | 'texts' | '_cells' ÏïàÏóê ÌÜ†ÌÅ∞ Î¶¨Ïä§Ìä∏ Ï°¥Ïû¨ Í∞ÄÎä•
    # - Î¨∏ÏûêÏó¥: Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©
    if isinstance(line, (list, tuple)):
        parts = [_tok_text(t) for t in line]
        parts = [p for p in parts if p]
        return " | ".join(parts) if parts else ""
    if isinstance(line, dict):
        for key in ("tokens", "_tokens", "line", "texts", "_cells"):
            v = line.get(key)
            if isinstance(v, (list, tuple)):
                parts = [_tok_text(t) for t in v]
                parts = [p for p in parts if p]
                return " | ".join(parts) if parts else ""
        # Îã®Ïùº ÌÖçÏä§Ìä∏ ÌïÑÎìúÍ∞Ä ÏûàÎäî Í≤ΩÏö∞
        if isinstance(line.get("text"), str):
            return line.get("text", "").strip()
        # Í∏∞ÌÉÄ Î¨∏ÏûêÏó¥ Í∞íÎßå Î™®ÏïÑÏÑú Î≥¥Ïó¨Ï£ºÍ∏∞(Ìè¥Î∞±)
        parts = [str(v).strip() for v in line.values() if isinstance(v, str)]
        return " | ".join(parts) if parts else str(line)
    # Í∏∞ÌÉÄ ÌÉÄÏûÖ Ìè¥Î∞±
    return _tok_text(line)

print("\nüßµ ÎùºÏù∏Î≥Ñ ÌÖçÏä§Ìä∏(ÌååÏù¥ÌîÑ Íµ¨Î∂Ñ):")
for idx, line in enumerate(ocr_lined_data, 1):
    try:
        s = _line_to_pipe(line)
    except Exception:
        s = str(line)
    print(f"  {idx}. {s}")
````

````python
# ÏµúÏ¢Ö Ï†ÄÏû• Î°úÏßÅÏùÑ Ìï®ÏàòÎ°ú Í∞êÏã∏Í≥†, OUTPUT_DIRÎ•º Ïù∏ÏûêÎ°ú Î∞õÏïÑ Ïã§ÌñâÌï©ÎãàÎã§.
from pathlib import Path
from datetime import datetime
import difflib
from typing import List, Tuple
import time


def run_final_save(OUTPUT_DIR, img_path: str, intermediates: dict, start_time: float):
    """
    ÏµúÏ¢Ö Í≤∞Í≥º Ï†ÄÏû• Ìï®Ïàò.
    - OUTPUT_DIR: Í≤∞Í≥º ÌååÏùº(.txt)ÏùÑ Ï†ÄÏû•Ìï† ÎîîÎ†âÌÑ∞Î¶¨(Path ÎòêÎäî str)
    - img_path: ÌòÑÏû¨ Ï≤òÎ¶¨ Ï§ëÏù∏ Ïù¥ÎØ∏ÏßÄÏùò Í≤ΩÎ°ú(str)
    - intermediates: extractor.extract_from_lines(..., return_intermediates=True) Í∞Ä Î∞òÌôòÌïú Ï§ëÍ∞Ñ ÏÇ∞Ï∂úÎ¨º(dict)
    - start_time: Ï≤òÎ¶¨ ÏãúÏûë ÏãúÍ∞Å (time.time() Í∞í). Ïù¥ ÏãúÍ∞ÅÏúºÎ°úÎ∂ÄÌÑ∞ Processing time ÏùÑ Í≥ÑÏÇ∞.

    ÎÇ¥Î∂ÄÏóêÏÑú ÏïÑÎûòÎ•º Íµ¨ÏÑ±ÌïòÏó¨ Ï†ÄÏû•Ìï©ÎãàÎã§:
      - step12_output = extractor.debug_step12(intermediates)
      - step13_output = extractor.debug_step13(intermediates)

    Ï†ÑÏ†ú(Í∏ÄÎ°úÎ≤å):
      - STEP12_MARKERS: List[str] (Step 12 Íµ¨Í∞ÑÏùò ÏãúÏûë ÏãùÎ≥ÑÏûêÎì§)
      - extractor: LabTableExtractor (debug_step12/13 Ï†úÍ≥µ)
    """
    # ÏïàÏ†Ñ Î≥ÄÌôò Î∞è Í∏∞Î≥∏ ÏÑ§Ï†ï
    output_dir = Path(OUTPUT_DIR).resolve()
    output_dir.mkdir(parents=True, exist_ok=True)

    base = Path(img_path).name  # ÏõêÎ≥∏ ÌôïÏû•Ïûê Ìè¨Ìï® ÌååÏùºÎ™Ö
    filename = f"{base}.txt"

    # --- Ïú†Ìã∏ Ìï®ÏàòÎì§ ---
    def has_step12_header(text: str, markers: List[str]) -> bool:
        t = (text or "").lstrip()
        for m in (markers or []):
            if t.startswith(m):
                return True
        # ÎÑêÎ¶¨ Ïì∞Îäî Í∏∞Î≥∏ Ìó§ÎçîÎèÑ Î∞©Ïñ¥Ï†ÅÏúºÎ°ú ÌôïÏù∏
        default_mark = "üßæ Step 12: Ìè¨Ìï®/Ï†úÏô∏ tests Ï†ÑÏ≤¥ (ÎùºÏù∏ Ïàú)"
        return t.startswith(default_mark)

    def build_header(img_path_str: str) -> str:
        now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        try:
            elapsed = float(time.time() - float(start_time))
        except Exception:
            elapsed = 0.0
        mm = int(elapsed // 60)
        ss = elapsed % 60
        image_name = Path(img_path_str).name
        header = [
            f"=== Output for {image_name} ===",
            f"Original image: {img_path_str}",
            f"Generated on: {now} (Processing time: {mm:02d}:{ss:05.2f})",
            ""
        ]
        return "\n".join(header)

    def build_body(step12: str, step13: str) -> str:
        lines: List[str] = []
        # Step 12 Ìó§ÎçîÍ∞Ä Ïù¥ÎØ∏ ÏûàÏúºÎ©¥ Ï§ëÎ≥µ Ï∂îÍ∞ÄÌïòÏßÄ ÏïäÏùå
        if not has_step12_header(step12, globals().get("STEP12_MARKERS", [])):
            lines.append("üßæ Step 12: Ìè¨Ìï®/Ï†úÏô∏ tests Ï†ÑÏ≤¥ (ÎùºÏù∏ Ïàú)")
        lines.append(step12.rstrip())
        lines.append("")
        # Step 13 ÏÑπÏÖò
        lines.append("üßæ Step 13: ÏµúÏ¢Ö ÌÖåÏù¥Î∏î")
        lines.append(step13.rstrip())
        return "\n".join(lines)

    def extract_step12_section(full_text: str) -> str:
        """Ï†ÄÏû•Îêú Ï†ÑÏ≤¥ ÌÖçÏä§Ìä∏ÏóêÏÑú Step 12 Íµ¨Í∞ÑÎßå Ï∂îÏ∂úÌïúÎã§."""
        if not full_text:
            return ""
        text = full_text
        # ÏãúÏûë ÎßàÏª§
        start_markers = globals().get("STEP12_MARKERS", [])
        default_start = "üßæ Step 12: Ìè¨Ìï®/Ï†úÏô∏ tests Ï†ÑÏ≤¥ (ÎùºÏù∏ Ïàú)"
        # ÎÅù ÎßàÏª§(= Step 13 ÏãúÏûë)
        end_marker = "üßæ Step 13: ÏµúÏ¢Ö ÌÖåÏù¥Î∏î"

        # ÏãúÏûë ÏúÑÏπò Ï∞æÍ∏∞
        start_idx = -1
        for m in start_markers + [default_start]:
            i = text.find(m)
            if i != -1:
                start_idx = i
                break
        if start_idx == -1:
            return ""  # Step 12 ÏÑπÏÖò ÏóÜÏùå

        # ÎÅù ÏúÑÏπò Ï∞æÍ∏∞
        end_idx = text.find(end_marker, start_idx + 1)
        if end_idx == -1:
            end_idx = len(text)

        return text[start_idx:end_idx].strip()

    def save_with_step12_guard(out_dir: Path, fname: str, header: str, body: str) -> Tuple[bool, Path]:
        target = (out_dir / fname)
        new_text = f"{header}\n{body}\n"
        print(f"[Ï†ÄÏû•] ÎåÄÏÉÅ: {target}")

        if target.exists():
            try:
                old_text = target.read_text(encoding="utf-8", errors="ignore")
            except Exception:
                old_text = ""
            # Step 12 ÏÑπÏÖòÎßå ÎπÑÍµê
            old_s12 = extract_step12_section(old_text)
            new_s12 = extract_step12_section(new_text)
            if old_s12 == new_s12:
                print("Step12 ÎèôÏùº ‚Üí Ï†ÄÏû• ÏßÑÌñâ(ÎçÆÏñ¥Ïì∞Í∏∞ ÌóàÏö©)")
            else:
                print("Step12 Ï∞®Ïù¥ Í∞êÏßÄ ‚Üí Ï†ÄÏû• Ï§ëÎã®, Step12 diff ÎØ∏Î¶¨Î≥¥Í∏∞")
                diff = difflib.unified_diff(
                    old_s12.splitlines(), new_s12.splitlines(),
                    fromfile="old_step12", tofile="new_step12", lineterm=""
                )
                preview = "\n".join(list(diff)[:80])  # ÎÑàÎ¨¥ Í∏∏Î©¥ ÏïûÎ∂ÄÎ∂ÑÎßå
                print(preview)
                return False, target

        # Ï†ÄÏû• ÏàòÌñâ
        try:
            target.write_text(new_text, encoding="utf-8")
            print("Ï†ÄÏû• ÏôÑÎ£å ‚úÖ")
            return True, target
        except Exception as e:
            print(f"Ï†ÄÏû• Ïã§Ìå®: {e}")
            return False, target

    # --- Ïã§ÌñâÎ∂Ä ---
    header_text = build_header(img_path)
    # intermediatesÎ•º Í∏∞Î∞òÏúºÎ°ú ÎîîÎ≤ÑÍ∑∏ ÌÖçÏä§Ìä∏ ÏÉùÏÑ±
    try:
        s12 = extractor.debug_step12(intermediates)
    except Exception as e:
        print(f"Step12 ÌÖçÏä§Ìä∏ ÏÉùÏÑ± Ïã§Ìå®: {e}")
        s12 = ""
    try:
        s13 = extractor.debug_step13(intermediates)
    except Exception as e:
        print(f"Step13 ÌÖçÏä§Ìä∏ ÏÉùÏÑ± Ïã§Ìå®: {e}")
        s13 = ""

    body_text = build_body(s12, s13)
    ok, saved_path = save_with_step12_guard(output_dir, filename, header_text, body_text)
    return ok, saved_path


# OUTPUT_DIR Î≥ÄÏàòÍ∞Ä ÏÇ¨Ï†ÑÏóê Ï†ïÏùòÎêòÏñ¥ ÏûàÏßÄ ÏïäÏúºÎ©¥ Í∏∞Î≥∏ Í≤ΩÎ°úÎ•º ÏÇ¨Ïö©
OUTPUT_DIR = Path("/home/aidan/work/meow-chat/notebooks/ocr/result").resolve()

img_paths = [
	os.path.abspath('notebooks/ocr/assets/images/20241106.jpg'),   # ---------- ÎÇ†ÏßúÍ∞Ä Ïó¨Îü¨Í∞ú
	os.path.abspath('notebooks/ocr/assets/images/20241107_2.jpg'),
	os.path.abspath('notebooks/ocr/assets/images/20241107.jpg'),
	os.path.abspath('notebooks/ocr/assets/images/20241121_2.png'),  # ---------- ÎÇ†Ïßú ÏóÜÎäîÍ≤É
	os.path.abspath('notebooks/ocr/assets/images/20241121.png'),   # ---------- ÏÇ¨ÏßÑÏúºÎ°ú ÏßùÏñ¥ÏÑú ÏÇêÎö§Ïñ¥ÏßÑÍ≤É
	os.path.abspath('notebooks/ocr/assets/images/20241123_2.png'),
	os.path.abspath('notebooks/ocr/assets/images/20241123.png'),   # ---------- Ïù∏ÏáÑÏùºÏûêÎßå ÏûàÏùå
	os.path.abspath('notebooks/ocr/assets/images/20241205_2.jpg'),
	os.path.abspath('notebooks/ocr/assets/images/20241205.jpg'),
	os.path.abspath('notebooks/ocr/assets/images/20241219.png'),    # ---------- ÏÇ¨ÏßÑÏúºÎ°ú ÏßùÏñ¥ÏÑú ÏÇêÎö§Ïñ¥ÏßÑÍ≤É
	os.path.abspath('notebooks/ocr/assets/images/20241221_2.jpg'),
	os.path.abspath('notebooks/ocr/assets/images/20241221.jpg'),
	os.path.abspath('notebooks/ocr/assets/images/20241224_2.png'),
	os.path.abspath('notebooks/ocr/assets/images/20241224.png'),
	os.path.abspath('notebooks/ocr/assets/images/20241231.png'),
	os.path.abspath('notebooks/ocr/assets/images/20250107.png'),    # ----------
	os.path.abspath('notebooks/ocr/assets/images/20250110.png'),    # ---------- Îç∞Ïù¥Î∏î Íµ¨Ï°∞ ÌäπÏù¥ÌïúÍ≤É. value Ïó¥ Î†àÏù¥Î∏îÏù¥ ÎÇ†ÏßúÏûÑ
	os.path.abspath('notebooks/ocr/assets/images/20250116_2.png'),  
	os.path.abspath('notebooks/ocr/assets/images/20250116_3.png'),  # ---------- blue Í∏ÄÏûê Ïù∏ÏãùÏù¥ Ïûò ÏïàÎê®
	os.path.abspath('notebooks/ocr/assets/images/20250116_4.png'), 
	os.path.abspath('notebooks/ocr/assets/images/20250116.png'), 
	os.path.abspath('notebooks/ocr/assets/images/20250119_2.png'),  # ---------- ÌùêÎ¶øÌïú Ïù¥ÎØ∏ÏßÄ Ìè∞Ìä∏ÏÉâÏÉÅ 
	os.path.abspath('notebooks/ocr/assets/images/20250119_3.png'), 
	os.path.abspath('notebooks/ocr/assets/images/20250119.png'),    # ---------- ÌùêÎ¶øÌïú Ïù¥ÎØ∏ÏßÄ
	os.path.abspath('notebooks/ocr/assets/images/20250125.jpeg'), # ----------   
	os.path.abspath('notebooks/ocr/assets/images/20250203.jpeg'), # ----------  
	os.path.abspath('notebooks/ocr/assets/images/20250207.jpeg'), # ----------   
	os.path.abspath('notebooks/ocr/assets/images/20250209.jpeg'), # ----------  
	os.path.abspath('notebooks/ocr/assets/images/20250216.jpeg'), # ----------  
	os.path.abspath('notebooks/ocr/assets/images/20250222.png'), # ----------  
	os.path.abspath('notebooks/ocr/assets/images/20250228.png'), # ----------   
	os.path.abspath('notebooks/ocr/assets/images/20250301.png'),
	os.path.abspath('notebooks/ocr/assets/images/20250306.jpeg'),
	os.path.abspath('notebooks/ocr/assets/images/20250309.jpeg'),
	os.path.abspath('notebooks/ocr/assets/images/20250315.png'),
	os.path.abspath('notebooks/ocr/assets/images/20250321.jpeg'),
	os.path.abspath('notebooks/ocr/assets/images/20250327.png'),
	os.path.abspath('notebooks/ocr/assets/images/20250329.png'),
	os.path.abspath('notebooks/ocr/assets/images/20250331.png'),
	os.path.abspath('notebooks/ocr/assets/images/20250410.png'),
	os.path.abspath('notebooks/ocr/assets/images/20250417.jpeg'),
]

print(f"[Ï†ÑÏ≤òÎ¶¨+OCR] Ï¥ù {len(img_paths)}Í∞ú Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨ ÏãúÏûë")
for p in img_paths:
    print(f"\n[ÏûëÏóÖ ÏãúÏûë] {p}")
    try:
        start_time = time.time()
        # Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏùÑ ÏùΩÍ≥† Ï†ÑÏ≤òÎ¶¨ÌïòÏó¨ Î∞îÏù¥Ìä∏Î°ú Î≥ÄÌôò
        with open(p, 'rb') as f:
            _raw_bytes = f.read()

        # ÌÅ¥ÎûòÏä§ Í∏∞Î∞ò Ï†ÑÏ≤òÎ¶¨ Ïã§Ìñâ (ÎîîÎ≤ÑÍ∑∏ Î°úÍ∑∏ Ìè¨Ìï®: Îç∞Ïä§ÌÅê Í∞ÅÎèÑ, Ïö©Îüâ Î≥ÄÌôî Îì±)
        _pre_bytes = ip.process_bytes(_raw_bytes, debug=True)
        print(f"‚úÖ ImagePreprocessor Î°ú Ï†ÑÏ≤òÎ¶¨ ÏôÑÎ£å: {p} (bytes: {len(_pre_bytes) if _pre_bytes else 'None'})")

        # OCR Ïã§Ìñâ
        ocr_result = ocr.run_ocr_from_bytes(_pre_bytes)
        if not ocr_result:
            print("‚ö†Ô∏è OCR Í≤∞Í≥ºÍ∞Ä ÎπÑÏñ¥ÏûàÏäµÎãàÎã§. Í±¥ÎÑàÎúÄ")
            continue

        # ÎùºÏù∏ Ï†ïÎ†¨ Î∞è Í∞í/Îã®ÏúÑ Î∂ÑÎ¶¨
        try:
            lined = lp.extract_and_group_lines(ocr_result[0])
        except Exception:
            # ÏùºÎ∂Ä OCR Î∞±ÏóîÎìúÍ∞Ä Î¶¨Ïä§Ìä∏ ÎåÄÏã† dictÎ•º Î∞òÌôòÌïòÎäî Í≤ΩÏö∞ Î∞©Ïñ¥
            first = ocr_result[0] if isinstance(ocr_result, (list, tuple)) else ocr_result
            lined = lp.extract_and_group_lines(first)

        # ÌÖåÏù¥Î∏î Ï∂îÏ∂ú (intermediates Ìè¨Ìï®)
        doc_result, intermediates = extractor.extract_from_lines(lined, return_intermediates=True)

        # Ï†ÄÏû•
        ok, path_saved = run_final_save(OUTPUT_DIR, p, intermediates, start_time)
        print(f"[Ï†ÄÏû• ÏÉÅÌÉú] {ok} -> {path_saved}")
    except Exception as e:
        print(f"‚ö†Ô∏è Ï≤òÎ¶¨ Ïã§Ìå®: {p} - {e}")
````
